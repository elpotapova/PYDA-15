```python
# перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Пункт 1 Пользователь по команде "p" может узнать владельца документа по его номеру

def which():
    doc_num = input('Введите номер документа: ')
    for num in documents:
        if doc_num in num['number']:
            print(f"Владелец документа: {num['name']}")
            break
    else:
        print('Документ не найден в базе')

# Пункт 2 Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

def where():
    doc_num = input('Введите номер документа: ')
    for num, direct in directories.items():
        if doc_num in direct:
            print(f'Документ хранится на полке: {num}')
            break
    else:
        print('Документ не найден в базе')

# Пункт 3 Пользователь по команде "l" может увидеть полную информацию по всем документам

def info():
    for inf in documents:
        phrase = '№: ' + inf['number'] + ', тип: ' + inf['type'] + ', владелец: ' + inf['name']
        for shelf in directories:
            sh_text = ', полка хранения: '
            if inf['number'] in directories[shelf]:
                sh_text = sh_text + shelf
                break
        print(phrase + sh_text)

# функция для добавления фразы про текущий перечень полок
def add_sh():
    sh = []
    for direct in directories:
        sh.append(int(direct))
    print('Текущий перечень полок: ' + str(sorted(sh)).strip('[]'))

# Пункт 4 Пользователь по команде "ads" может добавить новую полку

def new_shelf():
    num = input('Введите номер полки: ')
    if num in directories:
        print('Такая полка уже существует. ')
    elif num not in directories:
        directories[num] = []
        print('Полка добавлена. ')
    add_sh()

# Пункт 5 Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def del_shelf():
    shelf_num = input('Введите номер полки: ')
    for key, value in directories.items():
        if key in shelf_num:
            if value == []:
                directories.pop(key)
                print('Полка удалена. ')
                break
            else:
                print('На полке есть документы, удалите их перед удалением полки ')
                add_sh()
                
    if key not in shelf_num:
        print('Такая полка не существует.')
        
# сборная функция
        
def main():
    while True:
        command = input('Введите одну из команд: p, s, l, ads, ds. Для выхода команда q. Ваша команда: ')
        if command == 'p':
            which()
        elif command == 's':
            where()
        elif command == 'l':
            info()
        elif command == 'ads':
            new_shelf()
        elif command == 'ds':
            del_shelf()                    
        elif command == 'q':
            print('Exit')
            break
        else:
            print('Вы ввели некорректную команду, ознакомьтесь со списком команд.')
print(main())
```
