***Задание 1***

```python
import pandas as pd

def movies_class(rate):
    if rate <= 2:
        return 'Низкий рейтинг'
    elif rate < 4.5:
        return 'Средний рейтинг'
    elif rate >= 4.5:
        return 'Высокий рейтинг'
    else:
        return 'Некорректный рейтинг'
    
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
# соединяем 'movies' и 'ratings', удаляем лишние столбцы и используем ранее созданную функцию для получения нового столбца
class_ratings = ratings.groupby('movieId').mean().merge(movies, on='movieId', how='left')
class_ratings.drop(['userId', 'timestamp', 'genres'], axis='columns', inplace=True)
class_ratings['classification'] = class_ratings['rating'].apply(movies_class)
class_ratings.head()
```

***Задание 2***

```python
geo = pd.read_csv('keywords.csv')

geo_data = {'Центр': ['москв', 'тул', 'ярославл'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }


def geo_class(keywords):
    for key, value in geo_data.items():
        for city in value:
            if city in keywords:
                return key
        return 'undefined'
geo['region'] = geo['keyword'].apply(geo_class)
geo[geo['region'].isin(['Центр'])].head(15)
```
